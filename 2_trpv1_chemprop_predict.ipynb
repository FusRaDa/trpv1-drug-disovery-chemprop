````python
from chemprop.models.utils import load_model
from chemprop.models.model import MPNN
from chemprop import data, featurizers, models

import pandas as pd
import numpy as np
import torch
from lightning import pytorch as pl
from pathlib import Path
````

# Let's start predicting with these 5 novel compounds!

As well as 6 known compounds

````python
checkpoint_path = '/content/drive/MyDrive/Colab_Notebooks/TRPV1-drug-discovery-research/best.ckpt'
mpnn = MPNN.load_from_checkpoint(checkpoint_path)
mpnn
````

````python
test_data = [data.MoleculeDatapoint.from_smi(smi) for smi in nov_mols]
````

````python
featurizer = featurizers.SimpleMoleculeMolGraphFeaturizer()
test_dset = data.MoleculeDataset(test_data, featurizer=featurizer)
test_loader = data.build_dataloader(test_dset, shuffle=False)
````

````python
with torch.inference_mode():
    trainer = pl.Trainer(
        logger=None,
        enable_progress_bar=True,
        accelerator="cuda",
        devices=1
    )
    test_preds = trainer.predict(mpnn, test_loader)
````

````python
nov_mols_dict = {
        'compound_name': [
        'RTX',
        'novel_1',
        'novel_2',
        'novel_3',
        'novel_4',
        'novel_5',
        'Gingerol',
        'Shogaol',
        'Piperine',
        'Capsaicin',
        'CBD',
        'Caffiene',
    ],
    'SMILES': nov_mols,
}

nov_mols_df = pd.DataFrame(nov_mols_dict)
nov_mols_df
````

````python
test_preds = np.concatenate(test_preds, axis=0)
nov_mols_df['pred'] = test_preds
nov_mols_df
````

# Let's visualize these predictions

These predictions use known compounds that are known to interact with TRPV1 as well as Caffiene which does not have an interaction with TRPV1.

Remember that these prediction number are based on pChEMBL values which is intended to globally represent the activity of a compound based on their method of testing (IC50, pKb, Inhibition, etc...)

A pChEMBL value that is less than 5 is considered inactive and a value greater than 7 is considered highly active.

````python
smiles_list = list(nov_mols_df['SMILES'])
name_list = list(nov_mols_df['compound_name'])
pred_list = list(nov_mols_df['pred'])

all_mols = [Chem.MolFromSmiles(s) for s in smiles_list]
all_legends = []
for name, pred in zip(name_list, pred_list):
    all_legends.append(f"{name}: {pred:.2f}")

print("Known compounds and prediction of pChEMBL values:")
Draw.MolsToGridImage(all_mols, molsPerRow=6, subImgSize=(200, 200), legends=all_legends)
````

## What do these predicted pChEMBL values mean?
ChemProp only really aims to fit a chemical stucture (SMILES) to predict where it would fit in the training data. This is why it's not so suprising to see RTX and its similar compounds have close predicted pChEMBL values to each other.

### Let's compare the predicted value of RTX (CHEMBL17976) with the values of our training data.

````python
training_data = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/TRPV1-drug-discovery-research/trpv1_agonists_train_test_ids.csv')
rtx_vals = training_data[training_data['molecule_chembl_id'] == 'CHEMBL17976']
rtx_vals
````

````python
# Get average of these rows
rtx_vals_min = rtx_vals['pchembl_value'].min()
rtx_vals_max = rtx_vals['pchembl_value'].max()
rtx_vals_mean = rtx_vals['pchembl_value'].mean()
print(f"RTX min: {rtx_vals_min}, max: {rtx_vals_max}, mean: {rtx_vals_mean}")
````

### Now we know that our predicted value for RTX is 8.97 and the average of our training data is 9.2. What's the difference?

````python
pred_rtx_val = nov_mols_df[nov_mols_df['compound_name'] == 'RTX']['pred'].values[0]
deviation = abs(pred_rtx_val - rtx_vals_mean)
deviation
````

**Based on this observation, we get a 0.26 deviation between the predicted and training value, which is better than the models MAE of 0.42**
